```python
import pandas as pd

# Load Titanic training dataset
df = pd.read_csv("train.csv")

# Preview the data
df.head()
```

```python
# Step 4: Inspect column data types and missing values
df.info()
```

```python
# Step 4: Identification of Data Types (Based on .info() Output)
# ============================================================
# The Titanic dataset contains 891 observations and 12 variables with a mix
# of integer, floating-point, and object data types. Several variables require
# reclassification or transformation to align with their appropriate
# statistical data types.
#
# PassengerId is stored as an integer with no missing values, but it serves
# only as a unique identifier and does not provide predictive information.
#
# Survived is stored as an integer and represents a binary outcome variable
# (1 = survived, 0 = did not survive). This makes it suitable as the dependent
# variable for this classification task.
#
# Pclass is stored as an integer but represents passenger class levels (1st,
# 2nd, and 3rd). Although numeric in format, this variable is categorical in
# nature and ordinal because the classes follow a natural ranking.
#
# Sex and Embarked are object-type variables representing nominal categorical
# data with no inherent ordering.
#
# Age and Fare are floating-point variables representing continuous
# quantitative data. Age contains missing values,
# indicating that imputation will be required. Fare is fully populated and
# does not require missing-value treatment.
#
# SibSp and Parch are integer variables representing discrete quantitative
# counts of siblings/spouses and parents/children aboard the ship.
#
# Name and Ticket are object-type variables with high cardinality and do not
# encode structured or repeatable information useful for modeling.
#
# Cabin is also an object-type variable but contains a large proportion of
# missing values (only 204 non-null entries), making it unsuitable for direct
# inclusion without extensive preprocessing.
#
# Overall, this inspection shows the need for column removal, missing-value
# imputation, and proper reclassification before proceeding with modeling.
```

```python
# Step 5: Drop irrelevant or problematic columns
df = df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'])

# Fill missing Age values with median (robust to outliers)
df['Age'] = df['Age'].fillna(df['Age'].median())

# Fill missing Embarked values with mode
df['Embarked'] = df['Embarked'].fillna(df['Embarked'].mode()[0])

# Convert Pclass to categorical (ordinal)
df['Pclass'] = df['Pclass'].astype('category')

df.info()
```

```python
# Step 5: Data Transformation and Justification
# ============================================================
# After applying the necessary transformations, the dataset now contains
# 891 observations and 8 variables, all of which are fully populated with
# no missing values.
#
# PassengerId, Name, Ticket, and Cabin were removed from the dataset.
# PassengerId served only as a unique identifier, while Name and Ticket
# were high-cardinality text variables that do not encode structured
# predictive information. Cabin was removed due to a large proportion of
# missing values in the original dataset.
#
# Missing values in the Age variable were imputed using the median. Median
# imputation was chosen because it is robust to outliers and preserves the
# central tendency of the data. After imputation, Age now contains 891
# non-null observations and is stored as a continuous float variable.
#
# Missing values in the Embarked variable were imputed using the mode, which
# is appropriate for nominal categorical variables. As a result, Embarked
# now contains no missing values.
#
#
# These transformations ensure that each variable is represented using an
# appropriate statistical data type and that the dataset is clean and
# suitable for classification modeling.
```

```python
# Step 6:
# Dependent Variable (y): Survived
# This is a binary classification variable where:
# 1 = passenger survived
# 0 = passenger did not survive
```

```python
# Step 6: Identification of the Dependent Variable
# ============================================================
# The dependent variable for this analysis is 'Survived'.
# This variable is binary and indicates whether a passenger survived
# the Titanic disaster.
#
# A value of 1 represents that the passenger survived, while a value
# of 0 represents that the passenger did not survive.
#
# Because the outcome consists of two distinct categories, this
# variable is appropriate for use in a classification task.


```

```python
y = df['Survived']
```

```python
# Step 7: Identification of Independent Variables and Data Types
# ============================================================
# The independent variables consist of a combination of qualitative
# and quantitative features that describe passenger demographics,
# socioeconomic status, and family structure.
#
# Pclass is an ordinal categorical variable representing passenger
# class. Although numeric in its original format,
# it represents ordered categories rather than a continuous scale.
#
# Sex and Embarked are nominal categorical variables. These variables
# represent categories with no inherent ordering among their values.
#
# Age and Fare are continuous quantitative variables representing
# passenger age and ticket price, respectively.
#
# SibSp and Parch are discrete quantitative variables representing
# counts of siblings/spouses and parents/children aboard the ship.
#
# Together, these independent variables provide relevant information
# for predicting passenger survival.

```

```python
# Step 7: Independent variables
X = df.drop(columns=['Survived'])

# Identify variable groups
nominal_cols = ['Sex', 'Embarked']
ordinal_cols = ['Pclass']  # ordered categorical
numeric_cols = ['Age', 'Fare', 'SibSp', 'Parch']
```

```python
#  Step 8: Satistical Summary and Explanation


# Generate statistical summary for numeric variables
df.describe()
```

```python
# The statistical summary shows the central tendency, spread, and range of
# the quantitative variables in the dataset.
#
# Age has a mean in the late 20s and a wide range, indicating a diverse
# passenger age distribution. The median is slightly lower than the mean,
# suggesting mild right skew.
#
# Fare shows a much higher standard deviation relative to its mean, indicating
# strong right skew. This reflects the presence of a small number of very
# expensive tickets compared to many low-cost tickets.
#
# SibSp and Parch both have low means and medians of zero, indicating that most
# passengers traveled without immediate family members.
#
# The Survived variable has a mean below 0.5, indicating that fewer than half
# of the passengers survived. This suggests moderate class imbalance, which
# should be considered when evaluating model performance.
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Step 9: Train/Test Split


# Split the data into training and testing sets
# 80% training / 20% testing is used to provide sufficient data for model
# training while reserving enough unseen data for reliable evaluation.
# Stratification ensures that survival class proportions are preserved.

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
```

```python
# Step 10: Logistic Regression Model


# Preprocessing:
# - One-hot encode nominal categorical variables
# - Pass numeric and ordinal variables through unchanged

preprocessor = ColumnTransformer(
    transformers=[
        ('nominal', OneHotEncoder(handle_unknown='ignore'), nominal_cols),
        ('numeric', 'passthrough', numeric_cols + ordinal_cols)
    ]
)

logit_model = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('classifier', LogisticRegression(max_iter=1000))
])

logit_model.fit(X_train, y_train)
```

```python
# Logistic regression works by modeling the log-odds of survival as a linear
# combination of the independent variables. In simple terms, it estimates how
# each feature contributes to the probability that a passenger survived.
#
# Each coefficient shows how the log-odds of survival change when that variable
# increases by one unit, assuming all other variables stay the same. A positive
# coefficient means the feature increases the likelihood of survival, while a
# negative coefficient means it lowers the likelihood.
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, f1_score
```

```python
# Step 11: Logistic Regression Evaluation
# ============================================================

y_pred_logit = logit_model.predict(X_test)

cm_logit = confusion_matrix(y_test, y_pred_logit)
acc_logit = accuracy_score(y_test, y_pred_logit)
prec_logit = precision_score(y_test, y_pred_logit)
f1_logit = f1_score(y_test, y_pred_logit)

cm_logit, acc_logit, prec_logit, f1_logit
```

```python
# The logistic regression model performs solidly overall. The confusion matrix
# shows that most passengers are classified correctly, with relatively few
# misclassifications in both survival and non-survival categories.
#
# The accuracy is just over 80%, which indicates that the model does a good job
# predicting survival outcomes overall. Precision is also fairly strong,
# meaning that when the model predicts a passenger survived, it is usually
# correct. The F1 score suggests a reasonable balance between precision and
# recall, though the model still misses some survivors.
#
# Overall, logistic regression provides a strong and interpretable baseline
# model, and its performance is competitive with more complex approaches,
# making it a reliable choice for this dataset.

```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
```

```python
# Step 12: Neural Network Model
# ============================================================

# Standardization is applied to numeric variables to improve neural
# network convergence. A two-hidden-layer architecture is used to
# capture nonlinear relationships without excessive complexity.

nn_preprocessor = ColumnTransformer(
    transformers=[
        ('nominal', OneHotEncoder(handle_unknown='ignore'), nominal_cols),
        ('numeric', StandardScaler(), numeric_cols + ordinal_cols)
    ]
)

nn_model = Pipeline(steps=[
    ('preprocess', nn_preprocessor),
    ('classifier', MLPClassifier(
        hidden_layer_sizes=(64, 32),
        activation='relu',
        solver='adam',
        max_iter=500,
        random_state=42
    ))
])

nn_model.fit(X_train, y_train)
```

```python
# Step 13: Neural Network Evaluation
# ============================================================

y_pred_nn = nn_model.predict(X_test)

cm_nn = confusion_matrix(y_test, y_pred_nn)
acc_nn = accuracy_score(y_test, y_pred_nn)
prec_nn = precision_score(y_test, y_pred_nn)
f1_nn = f1_score(y_test, y_pred_nn)

cm_nn, acc_nn, prec_nn, f1_nn
```

```python
# The neural network performs reasonably well overall. The confusion matrix
# shows that the model correctly classifies most passengers, with more correct
# predictions than errors in both survival and non-survival cases.
#
# The accuracy of about 81% indicates that the model predicts the outcome
# correctly most of the time. Precision is fairly strong, meaning that when
# the model predicts a passenger survived, it is usually correct. The F1 score
# suggests a good balance between precision and recall, though there is still
# some room for improvement.
#
# Overall, the neural network captures patterns in the data better than a
# simple baseline, but it does not dramatically outperform simpler models,
# which suggests that the underlying relationships in the data may be fairly
# structured rather than highly complex.
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
# Step 14: Decision Tree Model
# ============================================================

tree_model = Pipeline(steps=[
    ('preprocess', preprocessor),
    ('classifier', DecisionTreeClassifier(
        max_depth=4,
        random_state=42
    ))
])

tree_model.fit(X_train, y_train)
```

```python
# This decision tree model uses a preprocessing pipeline that one-hot encodes
# categorical variables and passes numeric variables through unchanged before
# training the classifier.
#
# A maximum depth of 4 was chosen to keep the tree relatively simple and
# interpretable while still allowing it to capture meaningful patterns in the
# data. Limiting the depth also helps reduce overfitting, which can be an issue
# with decision trees if they grow too complex.
#
# Overall, this setup balances model interpretability with predictive power,
# making it useful for understanding how different features influence survival
# outcomes.
```

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
# Step 15: Decision Tree Visualization
# ============================================================

ohe = tree_model.named_steps['preprocess'].named_transformers_['nominal']
feature_names = list(ohe.get_feature_names_out(nominal_cols)) + numeric_cols + ordinal_cols

plt.figure(figsize=(20, 10))
plot_tree(
    tree_model.named_steps['classifier'],
    feature_names=feature_names,
    class_names=['Did Not Survive', 'Survived'],
    filled=True
)
plt.show()
```

```python
# Step 16: Decision Tree Evaluation

y_pred_tree = tree_model.predict(X_test)

cm_tree = confusion_matrix(y_test, y_pred_tree)
acc_tree = accuracy_score(y_test, y_pred_tree)
prec_tree = precision_score(y_test, y_pred_tree)
f1_tree = f1_score(y_test, y_pred_tree)

cm_tree, acc_tree, prec_tree, f1_tree
```

```python
# The decision tree shows mixed performance. The confusion matrix indicates
# that the model is very good at identifying passengers who did not survive,
# but it struggles more with correctly identifying survivors.
#
# The overall accuracy is just under 79%, which is slightly lower than the
# other models. Precision is quite high, meaning that when the model predicts
# survival, it is usually correct. However, the lower F1 score suggests that
# the model is missing a fair number of actual survivors, leading to weaker
# recall.
#
# This behavior is consistent with a shallow decision tree, which prioritizes
# simpler, more confident splits. While this improves interpretability and
# reduces overfitting, it can limit the model’s ability to capture more subtle
# patterns in the data.
```
